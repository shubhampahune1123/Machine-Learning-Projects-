```python
# UnderSampaling 
legit_sample=legit.sample(n=492)
```

```python
# splliting data into feature and target
x=ne_df.drop(columns="Class",axis=1)
y=ne_df["Class"]
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y , test_size=0.2,stratify=y,random_state=42)
```

```python
print()
```

```python
# model Training 
model=LogisticRegression()
```

```python
# Training the logistic regrestion Model
model.fit(x_train,y_train)
```

```python
# Model Evalution 

# ACuuracy Score  on Training data 
x_train_Pridiction=model.predict(x_train)
training_Data_Acuracy = accuracy_score(x_train_Pridiction,y_train)
```

```python
print("Accuracy on Traning data ",training_Data_Acuracy)
```

```python
# Accuracy on test data 
x_test_prediction=model.predict(x_test)
```

```python
test_data_Acuuracy =accuracy_score(x_test_prediction,y_test)
```

```python
print("Accuracy Score on test Data:" ,test_data_Acuuracy) 
```
